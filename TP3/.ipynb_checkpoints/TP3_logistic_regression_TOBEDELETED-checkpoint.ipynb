## 1. Logistic regression, IRLS algorithm.

### Preliminary question: the algorithm

#### Have a look at the function `regression_logistique.m` and locate the main steps of the algorithm you have been taught (see course).
You can comment the code in the cell below.

```python
def regression_logistique(X,t,Nitermax=20,eps_conv=1e-3):
    '''Entrees :
    X = [ones(N_train,1) x_train];
    t = class_train 
    Nitermax = nombre maximale d'itérations (20 par défaut)
    eps_conv = critère de convergence sur norm(w-w_old)/norm(w) ; 
    eps_conv=1e-3 par défaut
    
    Sorties : 
    w : vecteur des coefficients de régression logistique
   Niter : nombre d'itérations utilisées effectivement
   
   Fonction de régression logistique pour la classification binaire.
   
   Utilisation :
       Nitermax = 50
       eps_conv = 1e-4
       [w,Niter] = regression_logistique(X,t,Nitermax,eps_conv)
    '''
    N_train = X.shape[0]

    #initialisation : 1 pas de l'algorithme IRLS
    w = np.zeros((X.shape[1],))
    w_old = w 
    y = 1/2*np.ones((N_train,))
    R = np.diag(y*(1-y))   # diag(y_n(1-y_n))
    z = X.dot(w_old)-la.inv(R).dot(y-t)
    w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z)

    # boucle appliquant l'algorithme de Newton-Raphson
    Niter = 1
    while ( (la.norm(w-w_old)/la.norm(w)>eps_conv) & (Niter<Nitermax) ):
        Niter = Niter+1
        y = 1/(1+np.exp(-X.dot(w)))
        R = np.diag(y*(1-y))  
        w_old = w 
        z = X.dot(w_old)-la.inv(R).dot(y-t) 
        w = la.inv(X.T.dot(R).dot(X)).dot(X.T).dot(R).dot(z)
         
    return w, Niter
```

## 2. Application: handwritten digits recognition 5 & 6
We load 2 matrices which contain each a sequence of examples of 16x16 images 
of handwritten digits which are 5 and 6 here. Each line of the matrix
contains 256 pixel values coding for the gray level of a 16x16 image.

```python
train_5 = np.loadtxt('train_5.txt',delimiter=',')   # 556 samples
train_6 = np.loadtxt('train_6.txt',delimiter=',')   # 664 samples
```

### Examples of images


```python
# Digit 5
n=20;
I = np.reshape(train_5[n,:],(16,16))

plt.imshow(I,cmap='gray')
plt.show()
```

```python
# Digit 6
n=32;
I = reshape(train_6[n,:],(16,16))

plt.imshow(I,cmap='gray')
plt.show()
```

### Separating the training and test sets

We keep in the training set the 145 first images of 5s and the 200 first
images of 6s:


```python
''' 

1. Concatenate images of 5 and 6.
2. Generate array of 1s to put label in all observations.
3. Now put 0 in the label of all 145 first observations, beloinging to the images of number 5.
4. We end up with a separete array for the x's and the labels. x_train_brut that contains observations of the images of both numbers 5 and 6,
and class_train that contains their respective labels 0 and 1.

'''
x_train_brut = np.vstack((train_5[:145,:], train_6[:200,:]))
N_train = np.size(x_train_brut,axis=0)
class_train = np.ones((345,1))   # label 1 for digit 6
class_train[:145] = 0       # label 0 for digit 5

x_test_brut = np.vstack((train_5[145:,:], train_6[200:,:]))
N_test = np.size(train_5,axis=0)+np.size(train_6,axis=0)-N_train
class_test = np.ones((N_test,1))   # label 1 for digit 6
class_test[:(556-145)] = 0       # label 0 for digit 5
```

### 2. Most significant pixels

```python
# get standard deviation by pixel
std_by_pixel=np.std(x_train_brut, axis=0)
```

```python
# index of significant pixels
list_sig = np.where(std_by_pixel > 0.5)[0]

len(list_sig) #we get 173 significant pixels out of 256. So the images can be "summarized" just in 173 of them.
```

### 3. Binary image with most significant pixels

```python
# Create an array of 256 elements filled with 0 (just black pixels).
picture_sig = np.zeros(256, dtype=int)

# Replace with -1 the most significant pixels so that we can visualize in white the relevant ones.
picture_sig[list_sig] = 1
```

```python
# binary image with most significant pixels
I = reshape(picture_sig,(16,16))

plt.imshow(I,cmap='gray')
plt.show()
```

### 4. Define train dataset with significant pixels

```python
# we index all rows to maintain all samples, and filter columns of the matrix to keep only relevant indexes
x_train=x_train_brut[:,list_sig]
```

```python
# the resulting matrix has only 173 pixels as we discused previously
x_train.shape
```

### 5. Define test dataset with significant pixels

```python
# we index all rows to maintain all samples, and filter columns of the matrix to keep only relevant indexes
x_test=x_test_brut[:,list_sig]
```

```python
# the resulting matrix has only 173 pixels as we discused previously
x_test.shape
```

### 6. Calling newton raphson function to estimate the w's of the sigmoide function

```python
X = np.hstack((np.ones((N_train,1)),x_train))

# estimating the w's of our sigmoide function
w,Niter=regression_logistique(X,class_train.ravel(),13,1e-3)
```

### 7. Compute the decision function, and make predictions

```python
X_test=np.hstack((np.ones((N_test,1)),x_test))

y_sigmoide=1/(1+exp(-np.dot(X_test,w.reshape(174,1))))
```

```python
# computing thhe labels of the images in the test set, using as threshold 0.5
decision_vector_test=np.where(y_sigmoide>0.5,1,0)
```

### 8. Estimate the test error 

```python
# computing the error on the test set
print(f'error on the test set {(1-np.mean(decision_vector_test==class_test))*100}%')
```

### 9. Missclassified observations

```python
# get index of all images that were missclassified
missclassified_indexes = np.where(decision_vector_test != class_test)[0]
```

```python
# grab the pixels of the first missclassified image
missclassified_image=x_test_brut[missclassified_indexes[0],:]
```

```python
# plot missclassified image

print(f'below we can see the image of one missclassified observation. Which its true value is {5 if class_test[missclassified_indexes[0]][0] == 0 else 6}')

I = reshape(missclassified_image,(16,16))

plt.imshow(I,cmap='gray')
plt.show()

print()

print(f'However, the model said this image was {5 if decision_vector_test[missclassified_indexes[0]][0] == 0 else 6}')
```

```python
'''

Now, let's just do a loop to print all misclassified observations. 
So that we can check more of them quickly, and have more visual examples to
make conclusions.  

'''
for index in range(0,len(missclassified_indexes)):
    # grab the pixels of the first missclassified image
    missclassified_image=x_test_brut[missclassified_indexes[index],:]

    # plot missclassified image

    print(f'below we can see the image of one missclassified observation. Which its true value is {5 if class_test[missclassified_indexes[index]][0] == 0 else 6}')

    I = reshape(missclassified_image,(16,16))

    plt.imshow(I,cmap='gray')
    plt.show()

    print()

    print(f'However, the model said this image was {5 if decision_vector_test[missclassified_indexes[index]][0] == 0 else 6}')
    print()
    print()
```
